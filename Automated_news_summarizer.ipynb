## News Summarizer

### Import Packages

```python
from bs4 import BeautifulSoup
from requests import get
```

#### Creating a function to Extract only text form <p> Tags

```python
def get_only_text(url):
    '''
    return the title and the text of the article 
    at the specified url
    '''
    page = get(url)
    soup = BeautifulSoup(page.content, "lxml")
    text = " ".join(map(lambda p : p.text, soup.find_all("p")))
    title = " ".join(soup.title.stripped_strings)
    return title, text
```

#### calling the function with the url

```python
text = get_only_text("https://www.corriere.it/economia/tasse/26_gennaio_02/stipendi-come-cambiano-nel-2026-taglio-irpef-aumenti-e-sconti-fiscali-su-turni-festivi-e-buoni-pasto-ecco-la-nuova-busta-paga-083bfe8c-7b41-40f9-8850-d869ff95bxlk.shtml")
```

#### Summarization

```python
from sumy.summarizers.lex_rank import LexRankSummarizer
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
```

```python
parser = PlaintextParser.from_string(text[1], Tokenizer("italian"))
summarizer = LexRankSummarizer()
summary = summarizer(parser.document, 5)  
print("Title : ",text[0])
print("Summary : ")
for s in summary:
    print(s)
```

## Keywords

```python
import yake
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer("italian")
kw_extractor = yake.KeywordExtractor(lan="it", n=1, top=10)  # top 10 keywords
keywords = kw_extractor.extract_keywords(text[1])
for word in keywords:
    print(word[0])
```
